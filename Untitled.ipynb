```python
import pandas as pd
df = pd.read_csv("movie_dataset.csv")
print(df.head())
```

```python
print(df.columns)
```

```python
df=df[['genres','original_title','director']]
df=df.dropna()
print(df.head())
```

```python
df["combined_features"]=df["genres"]+" "+df["director"]
print(df[["original_title","combined_features"]].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert combined features (genres + director) into numerical form
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df["combined_features"])

print(tfidf_matrix.shape)  # Output: (number_of_movies, number_of_features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the similarity between movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

print(cosine_sim.shape)  # This should print (number_of_movies, number_of_movies)
```

```python
# Create a mapping of movie titles to their index
movie_indices = pd.Series(df.index, index=df["original_title"]).drop_duplicates()

# Function to get similar movie recommendations
def recommend_movies(title, num_recommendations=5):
    if title not in movie_indices:
        return "Movie not found in dataset."
    
    index = movie_indices[title]  # Get index of the movie
    similarity_scores = list(enumerate(cosine_sim[index]))  # Get similarity scores
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)  # Sort by similarity
    similarity_scores = similarity_scores[1:num_recommendations+1]  # Get top N recommendations
    
    recommended_movies = [df["original_title"].iloc[i[0]] for i in similarity_scores]  # Get movie titles
    return recommended_movies
```

```python
print(recommend_movies("Avengers: Age of Ultron"))
```
